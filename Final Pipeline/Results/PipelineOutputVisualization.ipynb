```python
for i in range(0,10):
  get_array=inputs[i]
  fig, axs = plt.subplots(nrows=1, ncols=11, figsize=(22, 2))
  for t in range (0,11):
    get_array_2=get_array[t]
    arr_transposed = np.transpose(get_array_2, axes=(1, 2, 0))
    arr_transposed = arr_transposed.astype(np.float32) / 255.0
    axs[t].imshow(arr_transposed)
    axs[t].axis('off')
```
